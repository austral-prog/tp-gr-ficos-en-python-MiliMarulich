```python
import matplotlib.pyplot as plt
import numpy as np
```

Gr√°fico 1

```python
years = [2021,2022,2023,2024,2025,2026,2027,2028,2029,2030]
market_size = [96,142,207,300,400,502,700,1400,1600,1800]

plt.plot(years, market_size, 'ro-', label='AI Market Size')

plt.title("Mercado global IA (2021 - 2030)")
plt.xlabel("A√±os")
plt.ylabel("Crecimiento")
plt.text(2028, 1400, "Gran salto de crecimiento", fontsize=10,  color = 'brown')
plt.grid(True)
plt.legend()
plt.xlim(2020, 2031)
plt.ylim(0, 2000)
plt.text(2025, 1800, "Punto importante", fontsize=10)
```

El gr√°fico de l√≠neas muestra el crecimiento proyectado del mercado global de inteligencia artificial entre 2021 y 2030, destacando su fuerte aumento.

Gr√°fico 2

```python
dias = ['Lunes','Martes','Mi√©rcoles', 'Jueves', 'Viernes', 'S√°bado', 'Domingo']
temperaturas = [18, 10, 15, 10, 5, 8, 9]

plt.bar(dias, temperaturas, color='mediumpurple') 
plt.title("Temperatura esta semana")
plt.xlabel("D√≠as")
plt.ylabel("Temperatura (¬∞C)") 
```

Gr√°fico 3

```python

plt.title("Incremento de horas que estudi√© durante el mes.")
plt.plot(np.random.random(7).cumsum(), color='orange')
plt.text(6, 4, "Por ex√°menes")

```

Este gr√°fico demuestra que a lo largo de este mes, las √∫ltimas semanas estudi√© a√∫n m√°s debido a los parciales. 

Gr√°fico 4 

```python
dias = ['Lunes', 'Martes', 'Mi√©rcoles', 'Jueves', 'Viernes', 'S√°bado', 'Domingo']
horas = [3, 3, 4, 4, 2, 5, 4]

plt.plot(range(len(dias)), horas, 'ro-', label='Tiempo en pantalla')

plt.title("Tiempo en pantalla de mi celular esta √∫ltima semana.")
plt.xlabel("D√≠as")
plt.ylabel("Horas")

plt.text(5, 5, "Aumento de tiempo.", fontsize=10, color='orange')  

plt.xticks(range(len(dias)), dias)

plt.grid(True)
plt.legend()
plt.ylim(0, 6)
```

Este gr√°fico demuestra el tiempo que permanec√≠ en pantalla a lo largo de la semana
