```python
import matplotlib.pyplot as plt
import numpy as np
```

# Gráfico 1

```python
years = [2021,2022,2023,2024,2025,2026,2027,2028,2029,2030]
market_size = [96,142,207,300,400,502,700,1400,1600,1800]

plt.plot(years, market_size, 'ro-', label='AI Market Size')

plt.title("Mercado global IA (2021 - 2030)")
plt.xlabel("Años")
plt.ylabel("Crecimiento")
plt.text(2028, 1400, "Gran salto de crecimiento", fontsize=10,  color = 'brown')
plt.grid(True)
plt.legend()
plt.xlim(2020, 2031)
plt.ylim(0, 2000)
plt.text(2025, 1800, "Punto importante", fontsize=10)
```

## El gráfico de líneas muestra el crecimiento proyectado del mercado global de inteligencia artificial entre 2021 y 2030, destacando su fuerte aumento.

# Gráfico 2

```python
dias = ['Lunes','Martes','Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']
temperaturas = [18, 10, 15, 10, 5, 8, 9]

plt.bar(dias, temperaturas, color='mediumpurple') 
plt.title("Temperatura esta semana")
plt.xlabel("Días")
plt.ylabel("Temperatura (°C)") 
```

## Este grafico demuestra como fue variando la temperatura a lo largo de la semana.

# Gráfico 3

```python

plt.title("Incremento de horas que estudie durante el mes.")
plt.plot(np.random.random(7).cumsum(), color='orange')
plt.text(6, 4, "Por examenes")

```

## Este grafico demuestra que a lo largo de este mes, las ultimas semanas estudie aun mas debido a los parciales. 

# Gráfico 4 

```python
dias = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']
horas = [3, 3, 4, 4, 2, 5, 4]

plt.plot(range(len(dias)), horas, 'ro-', label='Tiempo en pantalla')

plt.title("Tiempo en pantalla de mi celular esta última semana.")
plt.xlabel("Días")
plt.ylabel("Horas")

plt.text(5, 5, "Aumento de tiempo.", fontsize=10, color='orange')  

plt.xticks(range(len(dias)), dias)

plt.grid(True)
plt.legend()
plt.ylim(0, 6)
```

## Este grafico demuestra el tiempo que permaneci en pantalla a lo largo de la semana
